#### **DATA CLEANING:The Numbers Database**


```python
#Loading the Data and important librarioes.
import numpy as np
import pandas as pd
import gzip

tndb = pd.read_csv("../data/zippedData/tn.movie_budgets.csv.gz", compression= "gzip")
tndb.head()
```

```python
#CLEANING THE DATA:
#Cleaning the column names lowercase and replace spaces with underscores and remove parathenses.
tndb.columns = tndb.columns.str.strip().str.lower().str.replace(" ","_").str.replace("(","").str.replace(")","")
tndb.head()
tndb.dtypes
```

```python
#CLEAN DATES:
#Convert release date to datetime formart.
tndb["release_date"]= pd.to_datetime(tndb["release_date"], errors = "coerce")

#Extract year for analysis:
tndb["release_year"]= tndb["release_date"].dt.year
tndb.head()
```

```python
#CLEAN MOVIE TITLES:
#Remove extra whitespace from movie titles.
tndb["movie"] = tndb["movie"].str.strip()
tndb.head()
```

```python
#CLEAN FINACIAL DATA:
#Remove $ sign and convert to numeric.
tndb["production_budget"]= tndb["production_budget"].str.replace("$","").str.replace(",","")
tndb["production_budget"] = pd.to_numeric(tndb["production_budget"], errors="coerce")


tndb["domestic_gross"]= tndb["domestic_gross"].str.replace("$","").str.replace(",","")
tndb["domestic_gross"] = pd.to_numeric(tndb["domestic_gross"], errors="coerce")


tndb["worldwide_gross"]= tndb["worldwide_gross"].str.replace("$","").str.replace(",","")
tndb["worldwide_gross"] = pd.to_numeric(tndb["worldwide_gross"], errors="coerce")

tndb.head()
```

```python
#CREATE NEW COLUMNS:
#Calcuate profit
tndb["profit"]= tndb["worldwide_gross"]- tndb["production_budget"]

#Calculate return on investement(ROI)
tndb["roi"] = (tndb["profit"] / tndb["production_budget"] * 100)

#Calulate profit margin
tndb["foreign_gross"]= tndb["worldwide_gross"]- tndb["domestic_gross"]

tndb.head()
```

```python
#HANDLE DUPLICATES AND REMOVE MISSING VALUES:
#Sort by worldwide_gross and keep highest grosssing movie for each title.
tndb = tndb.sort_values("worldwide_gross", ascending= False)
tndb = tndb.drop_duplicates(subset= "movie", keep = "first")

#Drop rows where budget or worldwide_gross is missing
tndb = tndb.dropna(subset=["production_budget", "worldwide_gross"])
tndb.head()
```

```python
##FINAL CHECK OF CLEANED DATA:
#View cleaned data
tndb.head()
#Check for any remaining missing values
tndb.isnull().sum()
#Check for any remaining duplicates
tndb.duplicated(subset="movie").sum()
```

### **Question one: What is the optimal budget range for maximum return on investment (ROI)?**

To answer this question I:
- Filtered data to only movies with complete budget and roi information.
- Then created budget categories(budget bins and budget labels) to make it easier to see patterns in ranges rather than individual budget amounts.
- Calculated average ROI for each budget range so as to easily tell which budget range gives the best returns.
- Identified the best budget range
- Performed a statistical test to determine whether the best range significantly better than others. This was done to confirm the difference isnt just random i.e. if the p value <0.05 the difference becomes statistically signifcant(real, not chosen)


```python
#Filtered data to only movies with complete ROI with complete budget and ROI information: This was done because we cannot analyze ROI if we do not have the numbers.
q1_data = tndb[(tndb["production_budget"].notna()) & (tndb["roi"].notna())].copy()

#Created budget categories (range in dollars) for anlysis:
budget_bins = [0, 5_000_000, 10_000_000, 25_000_000, 50_000_000, 100_000_000, 500_000_000]
budget_lables = ["<$5M", "$5M-10M", "$10M-25M", "$25M-50M", "$50M-100M", ">$100M"]
q1_data["budget_range"]= pd.cut(q1_data["production_budget"],bins= budget_bins, labels = budget_lables)

#Budget bins are basically boundaries i.e. numerical cutoff points taht define where each category starts and ends. For example, the first bin (o-5M) includes all movies with a production budget greater than 0 and less than or equal to 5M
#Budget bins tell PANDAS to categorize each movie into one of these defined budget ranges based on its production budget.
#Budget labels on the other hand are the names we assigned to the budget ranges to amke them easier to understand and intepret during analysis.
#Thye basically tell PANDAS what to call each section.

#Create average ROI for each budget range: So as to determine which budget ranges have the highest ROI:
roi_summary = q1_data.groupby('budget_range', observed=True).agg({
    'roi': ['mean', 'median', 'std', 'count']
}).round(2)
roi_summary.columns = ['Mean_ROI', 'Median_ROI', 'Std_Dev', 'Film_Count']

#Created bar visualization to make it easier to spot the optimal range:
plt.figure(figsize=(12, 6))
budget_means = q1_data.groupby('budget_range', observed=True)['roi'].mean().sort_values(ascending=False)
plt.bar(range(len(budget_means)), budget_means.values, color='maroon', alpha=0.8)
plt.xlabel('Budget Range', fontsize=12)
plt.ylabel('Average ROI (%)', fontsize=12)
plt.title('Return on Investment by Budget Range', fontsize=14, fontweight='bold')
plt.xticks(range(len(budget_means)), budget_means.index, rotation=45)
plt.axhline(y=0, color='red', linestyle='--', alpha=0.5)  
plt.tight_layout()
plt.savefig('q1_roi_by_budget.png', dpi=300, bbox_inches='tight')
plt.show()

#Identified the optimal budget range based on the highest average ROI:
optimal_range = roi_summary["Mean_ROI"].idxmax()


#Performed a statistical test to determine which budget range is significantly better than others to confirm the difference is not just random:
# If the p value is less than 0.05 we can conclude that the differences in ROI between the budget ranges are statistically significant.
optimal_data = q1_data[q1_data['budget_range'] == optimal_range]['roi']
other_data = q1_data[q1_data['budget_range'] != optimal_range]['roi']
t_stat_q1, p_value_q1 = stats.ttest_ind(optimal_data, other_data, nan_policy='omit')

#Wrote a detailed summary of the findings:
print("RESULTS:")
print(roi_summary)
print(f"\nOptimal Budget Range: {optimal_range}")
print(f"Average ROI: {roi_summary.loc[optimal_range, 'Mean_ROI']:.2f}%")
print(f"Number of Films: {int(roi_summary.loc[optimal_range, 'Film_Count'])}")
print(f"Standard Deviation: {roi_summary.loc[optimal_range, 'Std_Dev']:.2f}%\n")
print(f"Statistical Test (t-test):")
print(f"t-statistic = {t_stat_q1:.3f}")
print(f"p-value = {p_value_q1:.4f}")
print(f"Significant? {'Yes (p < 0.05)' if p_value_q1 < 0.05 else 'No (p >= 0.05)'}\n")
print("CONCLUSION:")
print(f"The {optimal_range} budget range produces the highest average ROI of {roi_summary.loc[optimal_range, 'Mean_ROI']:.2f}%.")
print(f"This finding is {'statistically significant' if p_value_q1 < 0.05 else 'not statistically significant'}.\n")
print("RECOMMENDATION:")
print(f"YE Studios should target the {optimal_range} budget range for maximum profitability.")
print(f"Expected return: ~{roi_summary.loc[optimal_range, 'Mean_ROI']:.1f}% ROI")
```

### **Question two: Do independent(indie) studios achieve different average ratings compared to major studios, and what can we learn about the indie versus blockbuster strategy?**

To answer this question I:
- Filtered movies with both studios and ratings so as to compare indie vs major studio ratings.
- Defined which studios were "major" i.e. big players in the industry.
- Created a function to classify each studio as Indie or Major.
- Applied the classification to all movies and then removed unknowns for cleaner analysis.
- Calculated statistics for Indie vs Major studios to compare ratings and revenue between the two groups.
- Created 2 visualizations : The first, a box plot to compare ratings and how the full distributions and not just categories, the second, a bar chart comparing revenue to show the commercial success trade off.
- Performed a statistical test to determine whether the differences are significant to confirm the diffrence is real and not random.
- To supplement the statistical test, a Cohen's d test was done to calculate the effect size th show how meaningful the fifference was.(d<0.2=samll, d<0.5=medium and d>0.8=large effect)
- Wrote a detailed summary of the finings.

```python
#Filtered movies with both studio name and ratings so as to compare indie vs major studio ratings:
q2_data = df[df["studio"].notna() & df["averagerating"].notna()].copy()

#Defined which studios are major and which are indie based on the studio names so as to classify into two groups for comparison:
major_studios = ['Warner Bros.', 'WB', 'Universal', 'Paramount', 'Sony', 'Columbia','Disney', 'Walt Disney', '20th Century Fox', 'Fox', 'Lionsgate','MGM', 'DreamWorks', 'New Line']

#Created a function to classify whether a studio was major or indie:
def classify_studio(studio_name):
    if pd.isna(studio_name):
        return 'Unknown'
    studio_lower = str(studio_name).lower()
    for major in major_studios:
        if major.lower() in studio_lower:
            return 'Major'
    return 'Indie'

#Applied classification to all movies and removed unknowns for clearer analysis:
q2_data['studio_type'] = q2_data['studio'].apply(classify_studio)
q2_data = q2_data[q2_data['studio_type'] != 'Unknown']

#Calculated statistics for Indie vs Major studios to compare ratings and revenue between the two groups:
studio_stats = q2_data.groupby('studio_type').agg({
    'averagerating': ['mean', 'median', 'std', 'count'],
    'worldwide_gross': ['mean', 'median']
}).round(2)
studio_stats.columns = ['Mean_Rating', 'Median_Rating', 'Std_Rating', 'Count', 'Mean_Gross', 'Median_Gross']
print(studio_stats)

#Visualization 1: Box plot comparing ratings between Indie and Major Ratings to show full distributions not just averages
plt.figure(figsize=(10, 6))
sns.boxplot(data=q2_data, x='studio_type', y='averagerating', palette='Set2')
plt.xlabel('Studio Type', fontsize=12)
plt.ylabel('IMDB Rating', fontsize=12)
plt.title('Rating Comparison: Indie vs Major Studios', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('q2_ratings_comparison.png', dpi=300, bbox_inches='tight')
plt.show()

#Visuaization 2: Bar chart comparing revenue to show the commercial success trade-off between the two groups:
plt.figure(figsize=(10, 6))
revenue_means = q2_data.groupby('studio_type')['worldwide_gross'].mean() / 1_000_000
plt.bar(range(len(revenue_means)), revenue_means.values, color=['darkgrey', 'maroon'])
plt.xlabel('Studio Type', fontsize=12)
plt.ylabel('Average Worldwide Gross (Millions $)', fontsize=12)
plt.title('Revenue Comparison: Indie vs Major Studios', fontsize=14, fontweight='bold')
plt.xticks(range(len(revenue_means)), revenue_means.index)
plt.tight_layout()
plt.savefig('q2_revenue_comparison.png', dpi=300, bbox_inches='tight')
plt.show()

#Performed a statistical test to determine if the difference in ratings between Indie and Major studios is significant:
indie_ratings = q2_data[q2_data['studio_type'] == 'Indie']['averagerating']
major_ratings = q2_data[q2_data['studio_type'] == 'Major']['averagerating']
t_stat_q2, p_value_q2 = stats.ttest_ind(indie_ratings, major_ratings, nan_policy='omit')

#Supplemented the statistical test with a Cohen's d effect size calculation to understand the practical significance of the rating difference:
#Cohens is an important test because unlike p value which only tells us whether a difference is real, Cohens d tells us how big that difference is in a standardised way.
# d < 0.2 = small, d < 0.5 = medium, d > 0.8 = large effect
pooled_std = np.sqrt((indie_ratings.std()**2 + major_ratings.std()**2) / 2)
cohens_d = (indie_ratings.mean() - major_ratings.mean()) / pooled_std

#Wrote a detailed summary of the findings: To print and also saved to a text file for reporting:
# Print to console first
print("QUESTION 2: INDIE VS MAJOR STUDIOS")
print("="*70 + "\n")
print("RESULTS:")
print(studio_stats)
print("\nRATINGS:")
print(f"Indie average: {studio_stats.loc['Indie', 'Mean_Rating']:.2f}")
print(f"Major average: {studio_stats.loc['Major', 'Mean_Rating']:.2f}")
print(f"Difference: {abs(studio_stats.loc['Indie', 'Mean_Rating'] - studio_stats.loc['Major', 'Mean_Rating']):.2f} points\n")
print("REVENUE:")
print(f"Indie average: ${studio_stats.loc['Indie', 'Mean_Gross']:,.0f}")
print(f"Major average: ${studio_stats.loc['Major', 'Mean_Gross']:,.0f}")
print(f"Difference: ${abs(studio_stats.loc['Major', 'Mean_Gross'] - studio_stats.loc['Indie', 'Mean_Gross']):,.0f}\n")
print("STATISTICAL TEST:")
print(f"t-statistic = {t_stat_q2:.3f}")
print(f"p-value = {p_value_q2:.4f}")
print(f"Cohen's d = {cohens_d:.3f}")
print(f"Significant? {'Yes (p < 0.05)' if p_value_q2 < 0.05 else 'No (p >= 0.05)'}\n")
print("CONCLUSION:")
higher_rating = 'Indie' if studio_stats.loc['Indie', 'Mean_Rating'] > studio_stats.loc['Major', 'Mean_Rating'] else 'Major'
higher_revenue = 'Major' if studio_stats.loc['Major', 'Mean_Gross'] > studio_stats.loc['Indie', 'Mean_Gross'] else 'Indie'
print(f"{higher_rating} studios achieve higher ratings.")
print(f"{higher_revenue} studios generate higher box office revenue.")
print(f"This reveals a quality vs. revenue trade-off.\n")
print("RECOMMENDATION:")
print(f"For critical acclaim: Follow {higher_rating} studio approach")
print(f"For commercial success: Follow {higher_revenue} studio approach")
print(f"Hybrid strategy: Combine indie quality with major marketing")

# Then save to file
with open('q2_conclusion.txt', 'w') as f:
    f.write("QUESTION 2: INDIE VS MAJOR STUDIOS\n")
    f.write("="*70 + "\n\n")
    f.write("RESULTS:\n")
    f.write(f"{studio_stats}\n\n")
    f.write("RATINGS:\n")
    f.write(f"Indie average: {studio_stats.loc['Indie', 'Mean_Rating']:.2f}\n")
    f.write(f"Major average: {studio_stats.loc['Major', 'Mean_Rating']:.2f}\n")
    f.write(f"Difference: {abs(studio_stats.loc['Indie', 'Mean_Rating'] - studio_stats.loc['Major', 'Mean_Rating']):.2f} points\n\n")
    f.write("REVENUE:\n")
    f.write(f"Indie average: ${studio_stats.loc['Indie', 'Mean_Gross']:,.0f}\n")
    f.write(f"Major average: ${studio_stats.loc['Major', 'Mean_Gross']:,.0f}\n")
    f.write(f"Difference: ${abs(studio_stats.loc['Major', 'Mean_Gross'] - studio_stats.loc['Indie', 'Mean_Gross']):,.0f}\n\n")
    f.write("STATISTICAL TEST:\n")
    f.write(f"t-statistic = {t_stat_q2:.3f}\n")
    f.write(f"p-value = {p_value_q2:.4f}\n")
    f.write(f"Cohen's d = {cohens_d:.3f}\n")
    f.write(f"Significant? {'Yes (p < 0.05)' if p_value_q2 < 0.05 else 'No (p >= 0.05)'}\n\n")
    f.write("CONCLUSION:\n")
    f.write(f"{higher_rating} studios achieve higher ratings.\n")
    f.write(f"{higher_revenue} studios generate higher box office revenue.\n")
    f.write(f"This reveals a quality vs. revenue trade-off.\n\n")
    f.write("RECOMMENDATION:\n")
    f.write(f"For critical acclaim: Follow {higher_rating} studio approach\n")
    f.write(f"For commercial success: Follow {higher_revenue} studio approach\n")
    f.write(f"Hybrid strategy: Combine indie quality with major marketing\n")
```

To answer this question I:

#### **QUESTION 3: Financial Levels, Ratings, and Commercial Success.**

Analysis of 2,149 films across four financial tiers reveals weak correlation between budget and quality (r = 0.174), but strong correlation between budget and revenue (r = 0.793). **Conclusions:**

The low-budget tier's 389.96% ROI versus blockbusters' 247.32% demonstrates that investment efficiency peaks at lower budget levels, despite blockbusters achieving superior ratings (+0.87 points) and absolute revenues (7x higher gross).

```python
#Import my data for Tableau

import pandas as pd
import numpy as np

#I already have my dataframe from my analsysis so i will use it directly.

#Create Tableau ready dataset
tableau_data = df[['movie','production_budget','domestic_gross','foreign_gross', 'worldwide_gross','roi','profit','release_year','studio','averagerating','numvotes']].copy()

#Add studio type(Independent(Indie) vs Major) same classification from my nalaysis
major_studios = ['Warner Bros.', 'WB', 'Universal', 'Paramount', 'Sony', 'Columbia','Disney','Walt Disney', '20th Century Fox', 'Fox', 'Lionsgate','MGM', 'DreamWorks', 'New Line']

def classify_studio(studio_name):
    if pd.isna(studio_name):
        return 'Unknown'
    studio_lower = str(studio_name).lower()
    for major in major_studios:
        if major.lower() in studio_lower:
            return 'Major'
    return 'Indie'

tableau_data['studio_type'] = tableau_data['studio'].apply(classify_studio)

#Add budget range: same categories from my nalaysis for the first business question
budget_bins = [0, 5_000_000, 10_000_000, 25_000_000, 50_000_000, 100_000_000, 500_000_000]
budget_labels = ['<$5M', '$5-10M', '$10-25M', '$25-50M', '$50-100M', '>$100M']
tableau_data['budget_range'] = pd.cut(tableau_data['production_budget'],bins=budget_bins,labels=budget_labels)

#Add financial tiers: same categories from my analysis for the third business question
financial_bins = [0, 15_000_000, 50_000_000, 150_000_000, 500_000_000]
financial_labels = ['Low (<$15M)', 'Medium ($15-50M)', 'High ($50-150M)', 'Blockbuster (>$150M)']
tableau_data['financial_tier'] = pd.cut(tableau_data['production_budget'], bins=financial_bins, labels=financial_labels)

#Remove rows with nissing values in key columns to ensure clean visiaulizations in Tableau:
tableau_data = tableau_data.dropna(subset=['production_budget', 'worldwide_gross', 'roi'])

#Save for Tableau:
tableau_data.to_csv('tableau_movie_analysis.csv', index=False)

#Confirming that the dataset is ready for Tableau and what i am exporting:
print(f"Rows exported: {len(tableau_data)}")
print(f"\nColumns: {tableau_data.columns.tolist()}")
print(f"\nFirst few rows:")
print(tableau_data.head())
print(f"\nBudget range distribution:")
print(tableau_data['budget_range'].value_counts().sort_index())
```
